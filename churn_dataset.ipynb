**splitting the data**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_final.drop(['Churn_Yes'], axis = 1)
y = df_final['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
#used to scale the data
from sklearn.preprocessing import MinMaxScaler
```

```python
#scaling the data from 0 to 1
mms = MinMaxScaler(feature_range = (0, 1))
```

```python
X_train = mms.fit_transform(X_train)
X_test = mms.fit_transform(X_test)
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
#Here, we are creating a dictionary to compare the results of our various models
Results = {'Model':[], 'Accuracy':[],'Recall':[], 'Precision':[], 'F1':[]}
```

```python
Results = pd.DataFrame(Results)
Results.head()
```

```python
#I suppose this are all the logistic models that are commonly used
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
lr = LogisticRegression()
dc = DecisionTreeClassifier()
rf = RandomForestClassifier()
knn = KNeighborsClassifier(n_neighbors = 5)
svm = SVC()
```

```python
Results = pd.DataFrame(columns = ['Model', 'Accuracy', 'Recall', 'Precision','F1'])
model = [lr, dc, rf, knn, svm]
for models in model:
  models.fit(X_train, y_train)
  y_pred = models.predict(X_test)

  print('Model :', models)
  print('--------------------------------------------------------------')

  from sklearn.metrics import confusion_matrix
  from sklearn.metrics import classification_report#I donno what this classification report is...
  from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score

  print('confusion matrix : ', confusion_matrix(y_test, y_pred))
  print('classification report: ' ,classification_report(y_test, y_pred))
  print('Accuracy : ', round(accuracy_score(y_test,y_pred),2))
  print('recall: ', round(recall_score(y_test, y_pred),2))
  print('f1: ',round(f1_score(y_test, y_pred),2))
  print()#I am yet to understand why this space is here

  R = {'Model' :models,
       'Accuracy':round(accuracy_score(y_test,y_pred),2),
       'Recall':round(recall_score(y_test, y_pred),2),
       'Precision':round(precision_score(y_test, y_pred),2),
       'F1':round(f1_score(y_test,y_pred), 2)
       }
  Results = Results.append(R, ignore_index=True)









```

**Checking the best model among all of the above**

```python
#here, we are checking which model performed best
Results['Accuracy'] = Results['Accuracy'].astype(float)
Results['Precision'] = Results['Precision'].astype(float)
Results['Recall'] = Results['Recall'].astype(float)
Results['F1'] = Results['F1'].astype(float)

#.idmax() returns a Series with the index of the maximum value for each column
max_accuracy = Results['Accuracy'].idxmax()
max_precision = Results['Precision'].idxmax()
max_recall = Results['Recall'].idxmax()
max_f1 = Results['F1'].idxmax()

print('Best Accuracy Model:', Results.loc[max_accuracy]['Model'])
print('Best Precision Model:', Results.loc[max_precision]['Model'])
print('Best Recall Model:', Results.loc[max_recall]['Model'])
print('Best F1 Model:', Results.loc[max_f1]['Model'])
```
